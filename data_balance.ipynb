```python
# 디렉토리 경로 설정
directory = '/root/content2/train/images/'
cnt = 0
# 디렉토리 내의 파일 목록 가져오기
file_names = []
for file in os.listdir(directory):
    file_names.append(f'{file[:-4]}.json')

style = ['아우터', '상의', '하의', '원피스']

# 추출된 파일 이름 출력
for file_name in file_names:
    try:
        f = open('/root/content2/train/labels/'+file_name[:-4]+'txt', 'w')
        data = json.load(open('/root/trainning/label/'+file_name, encoding = 'utf-8'))
        width = data['이미지 정보']['이미지 너비']
        height = data['이미지 정보']['이미지 높이']

        for i in style:
            try:
                # 아우터 데이터
                if i == '아우터' and data['데이터셋 정보']['데이터셋 상세설명']['라벨링'][i] and data['데이터셋 정보']['데이터셋 상세설명']['렉트좌표'][i][0]:
                    label = 9
                    f.write(str(label) + ' ')

                    # 아우터 좌표 추출
                    for j in range(1, int(len(data['데이터셋 정보']['데이터셋 상세설명']['폴리곤좌표'][i][0])/2)+1):
                        try:
                            if data['데이터셋 정보']['데이터셋 상세설명']['폴리곤좌표'][i][0][f'X좌표{j}'] < 0:
                                x = 0.00000000001
                            elif data['데이터셋 정보']['데이터셋 상세설명']['폴리곤좌표'][i][0][f'X좌표{j}'] > width:
                                x = 0.99999999999
                            else:
                                x = data['데이터셋 정보']['데이터셋 상세설명']['폴리곤좌표'][i][0][f'X좌표{j}']/width

                            if data['데이터셋 정보']['데이터셋 상세설명']['폴리곤좌표'][i][0][f'Y좌표{j}'] < 0:
                                y = 0.00000000001
                            elif data['데이터셋 정보']['데이터셋 상세설명']['폴리곤좌표'][i][0][f'Y좌표{j}'] > height:
                                y = 0.99999999999
                            else:
                                y = data['데이터셋 정보']['데이터셋 상세설명']['폴리곤좌표'][i][0][f'Y좌표{j}']/height
                            f.write(str(x) + ' ')
                            f.write(str(y) + ' ')
                        except:
                            continue

                    f.write('\n')

                # 상의 데이터
                elif (i == '상의' and '소매기장' in data['데이터셋 정보']['데이터셋 상세설명']['라벨링'][i][0] and data['데이터셋 정보']['데이터셋 상세설명']['렉트좌표'][i][0]) and (data['데이터셋 정보']['데이터셋 상세설명']['라벨링']['아우터'][0] == {} or data['데이터셋 정보']['데이터셋 상세설명']['렉트좌표'][i][0] == {}):           

                    # 상의 라벨링
                    if data['데이터셋 정보']['데이터셋 상세설명']['라벨링'][i][0]['소매기장'] in ['민소매', '캡', '없음']:
                        label = 0
                    elif data['데이터셋 정보']['데이터셋 상세설명']['라벨링'][i][0]['소매기장'] == '반팔':
                        label = 1
                    elif data['데이터셋 정보']['데이터셋 상세설명']['라벨링'][i][0]['소매기장'] in ['7부소매', '긴팔']:
                        label = 2

                    f.write(str(label) + ' ')

                    # 상의 좌표 추출
                    for j in range(1, int(len(data['데이터셋 정보']['데이터셋 상세설명']['폴리곤좌표'][i][0])/2)+1):
                        try:
                            if data['데이터셋 정보']['데이터셋 상세설명']['폴리곤좌표'][i][0][f'X좌표{j}'] < 0:
                                x = 0.00000000001
                            elif data['데이터셋 정보']['데이터셋 상세설명']['폴리곤좌표'][i][0][f'X좌표{j}'] > width:
                                x = 0.99999999999
                            else:
                                x = data['데이터셋 정보']['데이터셋 상세설명']['폴리곤좌표'][i][0][f'X좌표{j}']/width

                            if data['데이터셋 정보']['데이터셋 상세설명']['폴리곤좌표'][i][0][f'Y좌표{j}'] < 0:
                                y = 0.00000000001
                            elif data['데이터셋 정보']['데이터셋 상세설명']['폴리곤좌표'][i][0][f'Y좌표{j}'] > height:
                                y = 0.99999999999
                            else:
                                y = data['데이터셋 정보']['데이터셋 상세설명']['폴리곤좌표'][i][0][f'Y좌표{j}']/height
                            f.write(str(x) + ' ')
                            f.write(str(y) + ' ')
                        except:
                            continue

                    f.write('\n')

                # 하의 데이터
                elif i == '하의' and '카테고리' in data['데이터셋 정보']['데이터셋 상세설명']['라벨링'][i][0] and '기장' in data['데이터셋 정보']['데이터셋 상세설명']['라벨링'][i][0] and data['데이터셋 정보']['데이터셋 상세설명']['렉트좌표'][i][0]:

                    # 하의 라벨링
                    if (data['데이터셋 정보']['데이터셋 상세설명']['라벨링'][i][0]['카테고리'] in ['청바지', '팬츠', '래깅스', '조거팬츠']) & (data['데이터셋 정보']['데이터셋 상세설명']['라벨링'][i][0]['기장'] in ['맥시', '발목', '미디']):
                        label = 3
                    elif (data['데이터셋 정보']['데이터셋 상세설명']['라벨링'][i][0]['카테고리'] in ['청바지', '팬츠', '래깅스', '조거팬츠']) & (data['데이터셋 정보']['데이터셋 상세설명']['라벨링'][i][0]['기장'] in ['미니', '니렝스']):
                        label = 4
                    elif (data['데이터셋 정보']['데이터셋 상세설명']['라벨링'][i][0]['카테고리'] == '스커트') & (data['데이터셋 정보']['데이터셋 상세설명']['라벨링'][i][0]['기장'] in ['맥시', '발목', '미디']):
                        label = 5
                    elif (data['데이터셋 정보']['데이터셋 상세설명']['라벨링'][i][0]['카테고리'] == '스커트') & (data['데이터셋 정보']['데이터셋 상세설명']['라벨링'][i][0]['기장'] in ['미니', '니렝스']):
                        label = 6

                    f.write(str(label) + ' ')

                    # 하의 좌표 추출
                    for j in range(1, int(len(data['데이터셋 정보']['데이터셋 상세설명']['폴리곤좌표'][i][0])/2)+1):
                        try:
                            if data['데이터셋 정보']['데이터셋 상세설명']['폴리곤좌표'][i][0][f'X좌표{j}'] < 0:
                                x = 0.00000000001
                            elif data['데이터셋 정보']['데이터셋 상세설명']['폴리곤좌표'][i][0][f'X좌표{j}'] > width:
                                x = 0.99999999999
                            else:
                                x = data['데이터셋 정보']['데이터셋 상세설명']['폴리곤좌표'][i][0][f'X좌표{j}']/width

                            if data['데이터셋 정보']['데이터셋 상세설명']['폴리곤좌표'][i][0][f'Y좌표{j}'] < 0:
                                y = 0.00000000001
                            elif data['데이터셋 정보']['데이터셋 상세설명']['폴리곤좌표'][i][0][f'Y좌표{j}'] > height:
                                y = 0.99999999999
                            else:
                                y = data['데이터셋 정보']['데이터셋 상세설명']['폴리곤좌표'][i][0][f'Y좌표{j}']/height
                            f.write(str(x) + ' ')
                            f.write(str(y) + ' ')
                        except:
                            continue

                    f.write('\n')


                ### 원피스 데이터
                elif i == '원피스' and '기장' in data['데이터셋 정보']['데이터셋 상세설명']['라벨링'][i][0] and data['데이터셋 정보']['데이터셋 상세설명']['렉트좌표'][i][0]:

                    # 원피스 라벨링
                    if data['데이터셋 정보']['데이터셋 상세설명']['라벨링'][i][0]['기장'] in ['맥시', '발목', '미디']:
                        label = 7
                    elif data['데이터셋 정보']['데이터셋 상세설명']['라벨링'][i][0]['기장'] in ['미니', '니렝스']:
                        label = 8

                    f.write(str(label) + ' ')

                    # 원피스 좌표 추출
                    for j in range(1, int(len(data['데이터셋 정보']['데이터셋 상세설명']['폴리곤좌표'][i][0])/2)+1):
                        try:
                            if data['데이터셋 정보']['데이터셋 상세설명']['폴리곤좌표'][i][0][f'X좌표{j}'] < 0:
                                x = 0.00000000001
                            elif data['데이터셋 정보']['데이터셋 상세설명']['폴리곤좌표'][i][0][f'X좌표{j}'] > width:
                                x = 0.99999999999
                            else:
                                x = data['데이터셋 정보']['데이터셋 상세설명']['폴리곤좌표'][i][0][f'X좌표{j}']/width

                            if data['데이터셋 정보']['데이터셋 상세설명']['폴리곤좌표'][i][0][f'Y좌표{j}'] < 0:
                                y = 0.00000000001
                            elif data['데이터셋 정보']['데이터셋 상세설명']['폴리곤좌표'][i][0][f'Y좌표{j}'] > height:
                                y = 0.99999999999
                            else:
                                y = data['데이터셋 정보']['데이터셋 상세설명']['폴리곤좌표'][i][0][f'Y좌표{j}']/height
                            f.write(str(x) + ' ')
                            f.write(str(y) + ' ')
                        except:
                            continue
            except:
                continue
        f.close()
        
        # 좌표값이 없거나 좌표값이 4개 이하인 txt파일 삭제
        with open('/root/content2/train/labels/'+file_name[:-4]+'txt', 'r') as fr:
            lines = fr.readlines()
            for line in lines:
                if (line == '') or (len(line) <= 5):
                    fr.close()
                    os.remove('/root/content2/train/labels/'+file_name[:-4]+'txt')
        cnt += 1
        print(cnt, '/', len(file_names))
    except:
        continue       
```
