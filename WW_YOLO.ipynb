```python
import os
import shutil
```

```python
# 이미지 원상복귀
path_t = os.listdir('/root/tmp/train/images')

for i in path_t:
    if i.endswith('.jpg'):
        shutil.move(f'/root/tmp/train/images/{i}', f'/root/tmp/images/{i}')
        
path_v = os.listdir('/root/tmp/valid/images')

for i in path_v:
    if i.endswith('.jpg'):
        shutil.move(f'/root/tmp/valid/images/{i}', f'/root/tmp/images/{i}')
```

```python
# 라벨(txt) 원상복귀
path_t = os.listdir('/root/tmp/train/labels')

for i in path_t:
    if i.endswith('.txt'):
        shutil.move(f'/root/tmp/train/labels/{i}', f'/root/tmp/txts/{i}')
    elif i.endswith('.json'):
        shutil.move(f'/root/tmp/train/labels/{i}', f'/root/tmp/labels/{i}')
        
path_v = os.listdir('/root/tmp/valid/labels')

for i in path_v:
    if i.endswith('.txt'):
        shutil.move(f'/root/tmp/valid/labels/{i}', f'/root/tmp/txts/{i}')
    elif i.endswith('.json'):
        shutil.move(f'/root/tmp/valid/labels/{i}', f'/root/tmp/labels/{i}')
```

```python
# 이미지 train, valid로 이동
data_name = os.listdir('/root/tmp/images')
data_name.sort()

for i in data_name[:14000]:
    shutil.move(f'/root/tmp/images/{i}', f'/root/tmp/train/images/{i}')
    
for j in data_name[14001:]:
    shutil.move(f'/root/tmp/images/{j}', f'/root/tmp/valid/images/{j}')
```

```python
# 라벨(txt) train, valid로 이동
label_name = os.listdir('/root/tmp/txts')
label_name.sort()

for k in label_name[:14000]:
    shutil.move(f'/root/tmp/txts/{k}', f'/root/tmp/train/labels/{k}')
    
for l in label_name[14001:]:
    shutil.move(f'/root/tmp/txts/{l}', f'/root/tmp/valid/labels/{l}')
```

```python
# 좌표값/라벨값 없는 txt파일 조회, 삭제(train)
file_names =[]
result = []

for file in os.listdir('/root/tmp/train/labels'):
    file_names.append(file)

for file_name in file_names:
    with open('/root/tmp/train/labels/' + file_name, 'r') as file:
        lines = file.readlines()
        if lines == []:
            result.append(file_name)
            
for i in result:
    if i in os.listdir('/root/tmp/train/labels'):
        os.remove(f'/root/tmp/train/labels/{i}')
        
    elif f'{i[:-4]}.jpg' in os.listdir('/root/tmp/train/images'):
        os.remove(f'/root/tmp/train/images/{i[:-4]}.jpg')
        
print(result)
```

```python
# 좌표값/라벨값 없는 txt파일 조회, 삭제(valid)
file_names =[]
result = []

for file in os.listdir('/root/tmp/valid/labels'):
    file_names.append(file)

for file_name in file_names:
    with open('/root/tmp/valid/labels/' + file_name, 'r') as file:
        lines = file.readlines()
        if lines == []:
            result.append(file_name)
            
for i in result:
    if i in os.listdir('/root/tmp/valid/labels'):
        os.remove(f'/root/tmp/valid/labels/{i}')
        
    elif f'{i[:-4]}.jpg' in os.listdir('/root/tmp/valid/images'):
        os.remove(f'/root/tmp/valid/images/{i[:-4]}.jpg')
        
print(result)
```
